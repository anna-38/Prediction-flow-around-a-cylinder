Import libraries

```python
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
import statistics as st
```

```python
# insert your path
path='./drive/MyDrive/Colab_Notebooks/Datamining_Project/'
```

Insert the files names for which you want to compute the accuracy

```python
data_real = 'tot_pressure.csv'
data_pred = 'Y_pres_cnn.csv'
path_real = path + data_real
path_pred = path + data_pred
```

Read the files

```python
tot=pd.read_csv(path_real, index_col=['Points:0','Points:1','Points:2'])
n_steps=75 # put the time steps
tot=tot.T.iloc[800+n_steps:].T
```

```python
Y_pred = pd.read_csv(path_pred)
Y_pred = Y_pred.T
```

Define the metrics functions

```python
def mae(y_true, y_pred):
  return mean_absolute_error(np.array(y_true),np.array(y_pred))
```

```python
def mape(y_true, y_pred):
    y_true=np.array(y_true)
    y_pred=np.array(y_pred)
    mape = np.mean(abs((y_true-y_pred)/y_true))*100
    return mape
```

```python
def smape(y_true, y_pred):
    y_true=np.array(y_true)
    y_pred=np.array(y_pred)
    smape = np.mean((np.abs(y_pred - y_true) * 200/ (np.abs(y_pred) + np.abs(y_true))))
    return smape
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred)) ** 2))
```

Mean of accuracy calculated for each point

```python
acc_mape=[]
acc_smape=[]
acc_rmse=[]
acc_mae=[]
for i in range(len(tot)):
  acc_mape.append(mape(tot.iloc[i],Y_pred.iloc[i]))
  acc_smape.append(smape(tot.iloc[i],Y_pred.iloc[i]))
  acc_rmse.append(rmse(tot.iloc[i],Y_pred.iloc[i]))
  acc_mae.append(mae(tot.iloc[i],Y_pred.iloc[i]))
print(st.mean(acc_mape))
print(st.mean(acc_smape))
print(st.mean(acc_rmse))
print(st.mean(acc_mae))
```
